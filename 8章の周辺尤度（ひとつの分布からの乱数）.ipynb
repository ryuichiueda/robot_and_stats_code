```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=0,scale = 1,size=1000)
data = np.sort(data)
data_mean = np.mean(data)
data_stdev = np.std(data)

print(data_mean)
print(data_stdev)
```

```python
likelihood_sum = sum([np.log(norm.pdf(e,data_mean, data_stdev)) for e in data])
print(likelihood_sum)
```

```python
data2 = [data[0:500], data[501:]]
data_means = [ np.mean(e) for e in data2]
data_stdevs = [ np.std(e) for e in data2 ]

print(data_means)
print(data_stdevs)
```

```python
likelihood_sums = 0
for i in range(2):
    likelihood_sums += sum( [np.log(0.5*norm.pdf(e,data_means[i], data_stdevs[i])) for e in data2[i] ] )
print(likelihood_sums)
```

```python
x = np.arange(-3,3,0.01)

plt.plot(x,norm.pdf(x,data_mean,data_stdev) ,color='r')
plt.plot(x,norm.pdf(x,data_means[0],data_stdevs[0])/2 ,color='b')
plt.plot(x,norm.pdf(x,data_means[1],data_stdevs[1])/2 ,color='b')

```
